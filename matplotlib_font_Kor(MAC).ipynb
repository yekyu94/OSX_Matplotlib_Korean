## OSX matplotlib 한글설정

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
%matplotlib inline
rc('font', family='AppleGothic') # 그래프 폰트 설정부분
```

```python
data = np.random.randint(-10, 10, 10)
plt.plot(data, 'r')
plt.title('상단 제목')
plt.ylabel('y축 항목')
plt.xlabel('x축 항목')
```
